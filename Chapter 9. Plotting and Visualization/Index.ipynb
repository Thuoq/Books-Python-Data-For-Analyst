### Annotaions and Drawing on a Subplot

Ngoài các loại cốt truyện tiêu chuẩn, bạn có thể muốn vẽ chú thích cốt truyện của riêng mình, có thể bao gồm văn bản, mũi tên hoặc các hình dạng khác. Bạn có thể thêm chú thích và văn bản `text`,`arrow`,`annotate`. văn bản vẽ văn bản tại các tọa độ đã cho (x, y) trên ô với kiểu tùy chỉnh tùy chọn ”

```python
#ax.test(x,y,'Hello world',family='monospace',fontsize=10)
```

```python
from datetime import datetime
import pandas as pd
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
data = pd.read_csv('../examples/spx.csv',index_col=0,parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax,style='k--')
crisis_data = [
    (datetime(2007,10,11),'Peak of bull market'),
    (datetime(2008,3,12),'Bear Stearns Fails'),
    (datetime(2008,9,15),'Lehman Bankruptcy')
]
for date,label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
               xytext=(date,spx.asof(date)+ 255),
                
               arrowprops=dict(facecolor='black'
                               ,headwidth=4
                               ,width=2
                              , headlength=4),
               horizontalalignment='left')
ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1900])
ax.set_title('Important dates in the 2008-2009 financial crisis');
```

Có một số điểm quan trọng cần làm nổi bật trong biểu đồ này: phương thức ax.annotate có thể vẽ các nhãn tại các tọa độ x và y được chỉ định. Chúng tôi sử dụng các phương thức `set_xlim `và `set_ylim` để đặt ranh giới bắt đầu và kết thúc theo cách thủ công cho cốt truyện thay vì sử dụng mặc định của `matplotlib`. Cuối cùng, `ax.set_title` thêm tiêu đề chính vào cốt truyện.

Xem thư viện `matplotlib` trực tuyến để biết thêm nhiều ví dụ chú thích khác để học hỏi.

Vẽ hình đòi hỏi một số cẩn thận hơn. `matplotlib` có các đối tượng đại diện cho nhiều hình dạng thông thường, được gọi là các bản vá lỗi. Một số trong số này, như Hình chữ nhật và Hình tròn, được tìm thấy trong `matplotlib.pyplot`, nhưng tập hợp đầy đủ nằm trong `matplotlib.patches.`

### Histograms and Density Plots 

**A histogram** là 1 loại đồ thị cột cho phép hiển thị dữ liệu 1 cách tuỳ biến các dữ liệu được chia ra các `bins` rời rạc, nếu số lượng thùng càng nhiều thì độ dài của thùng lại càng cao. 


```python
tips['tip_pct'].plot.hist(bins=50)
```

**Density plots** là Loại biểu đồ có liên quan là biểu đồ mật độ, được hình thành bằng cách tính toán ước tính về phân bố xác suất liên tục có thể tạo ra dữ liệu quan sát.

```python
tips['tip_pct'].plot.density()
```

**Seaborn** làm cho biểu đồ `histograms` và  `density` thậm chí còn dễ dàng hơn thông qua phương pháp phân tán của nó, có thể vẽ biểu đồ cả biểu đồ và ước tính mật độ liên tục đồng thời. Ví dụ: hãy xem xét một phân phối hai phương thức bao gồm các lần rút ra từ hai phân phối chuẩn chuẩn khác nhau ”

```python
comp1 = np.random.normal(0,1,size=200)
comp2 = np.random.normal(10,2,size=200)
values = pd.Series(np.concatenate([comp1,comp2]))
sns.histplot(values,bins=100,color='k',kde=True, stat="density", linewidth=0);
```

Biểu đồ **POINT** hoặc biểu đồ **Scatter** có thể là một cách hữu ích để kiểm tra mối quan hệ giữa hai chuỗi dữ liệu một chiều.

```python
macro = pd.read_csv('../examples/macrodata.csv')
```

Sau đó, chúng tôi có thể sử dụng phương pháp hồi quy của seaborn, phương pháp này tạo ra một biểu đồ phân tán và phù hợp với một đường hồi quy tuyến tính”

## “Facet Grids and Categorical Data”


Còn về bộ dữ liệu trong đó chúng ta có các thứ nguyên nhóm bổ sung? Một cách để trực quan hóa dữ liệu với nhiều biến phân loại là `facet grid`. Seaborn có một funciton `factorplot` tích hợp hữu ích giúp đơn giản hóa việc tạo nhiều loại biểu đồ nhiều mặt

```python
sns.catplot(x='day',y='tip_pct',hue='time',col='smoker',kind='bar',
              data=tips[tips.tip_pct < 1])
```

Thay vì group bơi `'time'` và `different bar colors` trong 1 grrid, chúng ta cũng có thể mở rông facet grid bơ thêm 1 giá trị Thời gian


```python
sns.catplot(x='day',y='tip_pct',hue='time',row='time',col='smoker',kind='bar',
              data=tips[tips.tip_pct < 1])
```
